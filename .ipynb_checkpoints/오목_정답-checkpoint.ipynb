```python
import numpy as np

sum1 = 0
sum2 = 0
sum3 = 0
sum4 = 0


size = int(input())
plate = np.zeros([size,size])

while True:
    print('p1 차례')
    p1x = int(input())-1
    if p1x >= size or p1x < 0:
        print("cannot")
        break
    elif p1x == -1:
        print('out')
        break
    p1y = int(input())-1
    if p1y >= size or p1y < 0:
        print("cannot")
        break
    elif p1y == -1:
        print('out')
        break
    p1y = size - 1 - p1y
    plate[p1y][p1x] = 1
    
    for i in range(size):
        for j in range(size):
            for k in range(5):
                if j + k >= size:
                    continue
                else:
                    sum1 += plate[i][j+k]
                if sum1 == 5:
                        print('p1 victory')
                        break
                        

            for k in range(5):
                if i + k >= size:
                    continue
                else:
                    sum2 += plate[i+k][j]
                if sum2 == 5:
                    print('p1 victory')
                    break
               
                    
            for k in range(5):
                if i + k >= size or j + k >= size:
                    continue
                else:
                    sum3 += plate[i+k][j+k]
                if sum3 == 5:
                    print('p1 victory')
                    break


            for k in range(5):
                if i + k >= size or j - k < 0:
                    continue
                else:
                    sum4 += plate[i+k][j-k]
                if sum4 == 5:
                    print('p1 victory')
                    break
            sum1, sum2, sum3, sum4 = 0, 0, 0, 0


    print('p2 차례')
    p2x = int(input())-1
    if p1x >= size or p1x < 0:
        print("cannot")
        break
    elif p2x == -1:
        print('out')
        break
    p2y = int(input())-1
    if p1y >= size or p1y < 0:
        print("cannot")
        break
    elif p2y == -1:
        print('out')
        break
    p2y = size - 1 - p2y
    plate[p2y][p2x] = -1
        
    for i in range(size):
        for j in range(size):
            for k in range(5):
                if j + k >= size:
                    continue
                else:
                    sum1 += plate[i][j+k]
                if sum1 == -5:
                    print('p2 victory')
                    break
            for k in range(5):
                if i + k >= size:
                    continue
                else:
                    sum2 += plate[i+k][j]
                if sum2 == -5:
                    print('p2 victory')
                    break
            for k in range(5):
                if i + k >= size or j + k >= size:
                    continue
                else:
                    sum3 += plate[i+k][j+k]
                if sum3 == -5:
                    print('p2 victory')
                    break
            for k in range(5):
                if i + k >= size or j - k < 0:
                    continue
                else:
                    sum4 += plate[i+k][j-k]
                if sum4 == -5:
                    print('p2 victory')
                    break
            sum1, sum2, sum3, sum4 = 0, 0, 0, 0
```
