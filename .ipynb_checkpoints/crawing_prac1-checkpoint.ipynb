```python
import requests
from bs4 import BeautifulSoup
```

```python
#실습1 - HTML 문서 내에 ID가 toc인 태그내의 내용을 출력해주세요
URL_w = 'https://ko.wikipedia.org/wiki/위키백과'
```

```python
response_w = requests.get(URL_w)
response_w.status_code
```

```python
soup = BeautifulSoup(response_w.text, 'html.parser')
soup.find('title')
```

```python
result_w = soup.find('div')
print(result_w)
```

```python
result_w = soup.find(id='toc')
```
