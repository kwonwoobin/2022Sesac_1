```python

import numpy as np
size=int(input('판 크기를 입력해주세요: '))
print('0을 입력하면 게임이 종료됩니다.')
print('게임시작!')

arr1=np.zeros((size,size))
idx=1

sum1 = 0
sum2 = 0
sum3 = 0
sum4 = 0


while True:
    if idx==1:
        r, c = map(int,input().split()-1) #사용자는 1,2넣으면 컴은 0,1로 받아들이도록..
        print('플레이어{} 바둑돌 위치 x y:'.format(idx), r, c)
        
        if r<0 or c<0 :
            print(int(input("0이상의 수로 다시 입력하세요: ")
            break
            
        elif r == 0 or c == 0:
            print('0을 입력하였습니다. 게임을 종료합니다.')
            break
            

        if r==0:
             r+=4
        elif r==1:
            r+=2
        elif r==2:
             r=2
        elif r==3:
            r-=2
        elif r==4:
            r-=4
                    
                    
        if arr1[r][c] == 0:              
            arr1[r][c]=1
            for i in range(size):
                for j in range(size):
                    if arr1[i][j]: 
                        for z in range(5):
                            if j+z >= size:
                                continue
                            else:
                                sum1 += arr1[i][j+z]
                        for z in range(5):
                            if i+z >= size or j+z >= size:
                                continue
                            else:
                                sum2 += arr1[i+z][j+z]
                        for z in range(5):
                            if i-z >= size or i-z<=0:
                                continue
                            else:
                                sum3 += arr1[i-z][j]
                        for z in range(5):
                            if i-z <= 0 or i+z >= size:
                                continue
                            else:
                                sum4 += arr1[i-z][i+z]
                        
                                
                        if sum1 ==5 or sum2 ==5 or sum3== 5 or sum4==5:
                          winner = 1
                          print("winner: {}".format(winner))
                        else:
                            continue
                      sum1, sum2, sum3, sum4 = 0, 0, 0, 0
                            
    if idx==1:
        idx+=1
    elif idx==2:
        idx-=1


    if idx==2:
        r, c = map(int,input().split()-1) #사용자는 1,2넣으면 컴은 0,1로 받아들이도록..
        print('플레이어{} 바둑돌 위치 x y:'.format(idx), r, c)
        
        if r<0 or c<0 :
            print(int(input("0이상의 수로 다시 입력하세요: ")
            break
                      
        elif r == 0 or c == 0:
            print('0을 입력하였습니다. 게임을 종료합니다.')
            break
            

        if r==0:
             r+=4
        elif r==1:
            r+=2
        elif r==2:
             r=2
        elif r==3:
            r-=2
        elif r==4:
            r-=4
        
                    
        if arr1[r][c] == 0:              
            arr1[r][c]=-1
            for i in range(size):
                for j in range(size):
                    if arr1[i][j]: 
                        for z in range(5):
                            if j+z >= size:
                                continue
                            else:
                                sum1 += arr1[i][j+z]
                        for z in range(5):
                            if i+z >= size or j+z >= size:
                                continue
                            else:
                                sum2 += arr1[i+z][j+z]
                        for z in range(5):
                            if i-z >= size or i-z<=0:
                                continue
                            else:
                                sum3 += arr1[i-z][j]
                        for z in range(5):
                            if i-z <= 0 or i+z >= size:
                                continue
                            else:
                                sum4 += arr1[i-z][i+z]
                        
                                
                        if sum1 ==-5 or sum2 ==-5 or sum3== -5 or sum4==-5:
                          winner = 2
                          print("winner: {}".format(winner))
                        else:
                            continue
                      sum1, sum2, sum3, sum4 = 0, 0, 0, 0
                            
        
                
                
                
     #플레이어 구분 : c=(c+1)%2
            
    #int(input()-1)
```
